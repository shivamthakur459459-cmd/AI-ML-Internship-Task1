```python
# ================================
# Task 1: Understanding Dataset & Data Types
# Dataset: Titanic
# ================================

# Step 1: Import libraries
import pandas as pd
import numpy as np

# Step 2: Load Dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("Dataset Loaded Successfully!")

# Step 3: First and Last rows
print("\n--- First 5 Rows ---")
print(df.head())

print("\n--- Last 5 Rows ---")
print(df.tail())

# Step 4: Dataset Information
print("\n--- Dataset Info ---")
print(df.info())

# Step 5: Statistical Summary
print("\n--- Statistical Summary ---")
print(df.describe())

# Step 6: Identify Data Types Manually
numerical_columns = ['Age', 'Fare', 'SibSp', 'Parch']
categorical_columns = ['Sex', 'Embarked', 'Pclass']
binary_columns = ['Survived']

print("\nNumerical Columns:", numerical_columns)
print("Categorical Columns:", categorical_columns)
print("Binary Column:", binary_columns)

# Step 7: Unique values in categorical columns
print("\n--- Unique Values ---")
for col in categorical_columns:
    print(f"{col}: {df[col].unique()}")

# Step 8: Target Variable and Input Features
target_variable = 'Survived'
input_features = df.drop(columns=[target_variable]).columns.tolist()

print("\nTarget Variable:", target_variable)
print("Input Features:", input_features)

# Step 9: Missing Values
print("\n--- Missing Values ---")
print(df.isnull().sum())

# Step 10: Dataset Size
rows, cols = df.shape
print(f"\nDataset Size: {rows} rows and {cols} columns")

# Step 11: ML Suitability Observation
print("\n--- ML Readiness Observations ---")
print("""
1. Dataset is structured and well-defined.
2. Target variable is binary (Classification problem).
3. Missing values present in Age, Cabin, and Embarked columns.
4. Requires data cleaning before modeling.
5. Suitable for Machine Learning classification tasks.
""")
```
